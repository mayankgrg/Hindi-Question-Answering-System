```
import requests
r = requests.get('https://www.focusonlearn.com/mughal-vansh-gk-in-hindi/')
c = r.content
from bs4 import BeautifulSoup
soup = BeautifulSoup(c)
main_content = soup.find('div',attrs={'class':'entry-content'})
ques=main_content.find_all('p')
tree=[]
for row in ques:
  #question1.append(row.get_text()[3:])
  tree.append(row.get_text())
```

```
i=3
while i<63:
  question.append(tree[i])
  i+=2
```

```
i=4
while i< 64:
  answer.append(tree[i])
  i+=2
```

```
i=64
while i<105:
  question.append(tree[i])
  i+=2
```
